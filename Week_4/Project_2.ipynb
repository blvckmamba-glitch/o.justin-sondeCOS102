```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.dropna()
```

```python
survival_counts = df.groupby(['Sex', 'Survived']).size().unstack()

survival_counts.plot(kind='bar', stacked=False, color=['darkred', 'pink'])

plt.title("Survival Counts by Sex")
plt.xlabel("Sex")
plt.ylabel("Number of Passengers")
plt.legend(["Did Not Survive", "Survived"])
plt.show()
```

```python
Insight:
      Females have more survival rate than males
```

```python
ages = df['Age'].dropna()

plt.figure(figsize=(10,6))
plt.hist(ages, bins=30, color='darkred', edgecolor='black')
plt.title("Age Distribution of Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

Insight:
       Most passengers were young adults, especially in their 20s and 30s

```python
class_counts = df['Pclass'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(class_counts, 
        labels=class_counts.index, 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=['pink', 'darkred', 'darkblue'])

plt.title("Passenger Class Distribution")
plt.show()
```

Insight:
       The largest group of passengers was in 3rd class, followed by 2nd class, with the smallest group in 1st class.
