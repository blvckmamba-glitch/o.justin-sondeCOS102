```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/HP/Downloads/archive/AB_NYC_2019.csv")
df
```

```python
df.info()
```

```python
df.dropna()
```

```python
#Listings by Neighborhood
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="neighbourhood_group", order=df['neighbourhood_group'].value_counts().index)
plt.title("Number of Listings by Neighborhood Group")
plt.show()
```

Insight:
      Manhattan has the highest number of listings, showing it’s the most popular area.


```python
#Price Distribution
plt.figure(figsize=(8,4))
sns.histplot(df['price'], bins=50)
plt.title("Distribution of Airbnb Prices in NYC")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

Insight:
    Prices are skewed, with most affordable options under $500.


```python
#Availability vs. Price
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="price", y="availability_365", alpha=0.3)
plt.title("Price vs Availability (days per year)")
plt.xlabel("Price")
plt.ylabel("Availability (days)")
plt.show()
```

Insight:
     - Budget-friendly listings tend to be available year-round, while expensive ones are limited.
