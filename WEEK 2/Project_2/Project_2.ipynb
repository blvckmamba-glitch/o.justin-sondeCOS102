START

DISPLAY "Choose equation to solve"
DISPLAY "1. Quadratic Equation"
DISPLAY "2. Cubic Equation"
DISPLAY "3. Quartic Equation"

INPUT choice

IF choice == 1 THEN
    INPUT a, b, c
    COMPUTE discriminant = b^2 - 4ac
    COMPUTE root1 = (-b + sqrt(discriminant)) / (2a)
    COMPUTE root2 = (-b - sqrt(discriminant)) / (2a)
    PRINT root1, root2

ELSE IF choice == 2 THEN
    INPUT a, b, c, d
    USE numerical method to find cubic roots
    PRINT roots

ELSE IF choice == 3 THEN
    INPUT a, b, c, d, e
    USE numerical method to find quartic roots
    PRINT roots

ELSE
    PRINT "Invalid option"

END

```python
import numpy as np

print("Choose the equation to solve")
print("1. Quadratic Equation")
print("2. Cubic Equation")
print("3. Quartic Equation")

choice = int(input("Enter your choice: "))

# Quadratic
if choice == 1:
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))

    discriminant = (b**2) - (4*a*c)

    root1 = (-b + discriminant**0.5) / (2*a)
    root2 = (-b - discriminant**0.5) / (2*a)

    print("Roots are:", root1, root2)

# Cubic
elif choice == 2:
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))

    coeff = [a, b, c, d]
    roots = np.roots(coeff)

    print("Roots are:", roots)

# Quartic
elif choice == 3:
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))
    e = float(input("Enter E: "))

    coeff = [a, b, c, d, e]
    roots = np.roots(coeff)

    print("Roots are:", roots)

else:
    print("Invalid choice")
```
